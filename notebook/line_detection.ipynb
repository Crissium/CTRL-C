```python
import numpy
import cv2
from pylsd import lsd
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('a.jpg')
plt.imshow(img[:,:,::-1])
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
segs = lsd(gray, scale=0.5)

plt.imshow(gray, cmap='gray')
plt.plot((segs[:,0],segs[:,2]), (segs[:,1],segs[:,3]))
plt.show()
```
